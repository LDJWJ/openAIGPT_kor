### API 호출 및 모델 목록 표시하기

### 사전 준비
 *  아래의 코드 실행을 위해서는 chatgpt.env의 파일 내용이 구글 코랩에 업로드 또는 생성되어 있어야 한다.
 * 사전에 OpenAI API를 발급받아야 한다. 만약 무료 버전이라면 일정 기간이 지나면 만료되므로 이후, 유료 결제 후 아래 내용을 진행해야 한다.

````python
import os
from openai import OpenAI

# chatgpt.env 파일에서 API_KEY와 ORG_ID의 값 읽어오기
with open("chatgpt.env") as env:
   for line in env:
       key, value = line.strip().split("=")
       os.environ[key] = value

# API key와 조직 ID 초기화
client = OpenAI(api_key  = os.environ.get("API_KEY"),
                organization  = os.environ.get("ORG_ID"))

# API 호출 및 모델 목록 표시
models = client.models.list()
print(models)
````

### 모델의 ID 출력

````python
import os
from openai import OpenAI

# chatgpt.env 파일에서 API_KEY와 ORG_ID의 값 읽어오기
with open("chatgpt.env") as env:
   for line in env:
       key, value = line.strip().split("=")
       os.environ[key] = value

# API key와 조직 ID 초기화
client = OpenAI(api_key  = os.environ.get("API_KEY"),
                organization  = os.environ.get("ORG_ID"))

# API 호출 및 모델 목록 표시
models = client.models.list()
for model in models.data:
    print(model.id)
````
